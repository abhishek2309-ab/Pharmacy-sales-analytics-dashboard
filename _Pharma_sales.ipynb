```python
# Convert Date column into datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Create Month column
df['Month'] = df['Date'].dt.month_name()
```

```python
df
```

```python
# Total Revenue
total_revenue = df['Revenue'].sum()
print(total_revenue)
```

```python
# Total Profit
total_profit = df['Profit'].sum()
print(total_profit)
```

```python
# Total Quantity Sold
total_quantity = df['Quantity'].sum()
print(total_quantity)
```

```python
# Average Revenue
average_revenue = df['Revenue'].mean()
print(average_revenue)
```

```python
#Top selling medicines.
top_medicines = (df.groupby('Medicine_Name')['Revenue'].sum().sort_values(ascending=False))
print(top_medicines.head())
```

```python
#monthly sales trend
monthly_sales = ( df.groupby(df['Date'].dt.month)['Revenue'].sum())
print(monthly_sales)
```

```python
df.to_csv("cleaned_pharmacy_sales.csv", index=False)
print("Cleaned dataset saved successfully")
```
